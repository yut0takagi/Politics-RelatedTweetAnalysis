```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.common.exceptions import ElementClickInterceptedException, NoSuchElementException
import time
import json
import datetime

# --- Selenium設定 ---
options = webdriver.ChromeOptions()
# options.add_argument("--headless")  # 必要なら無効化
driver = webdriver.Chrome(options=options)

# --- 政治記事一覧ページを開く ---
driver.get("https://news.yahoo.co.jp/categories/domestic")
time.sleep(2)

# --- 「政治」タブをクリック ---
try:
    driver.find_element(By.XPATH, '//button[text()="政治"]').click()
    time.sleep(2)
except Exception as e:
    print("政治ボタンが見つかりませんでした:", e)
    driver.quit()

for i in range(10):
    try:
        print(f"▶ {i+1} 回目クリック")
        load_more_button = driver.find_element(By.XPATH, '//button[text()="もっと見る"]')
        driver.execute_script("arguments[0].click();", load_more_button)  # JavaScriptで強制クリック
        time.sleep(2.5)  # 記事読み込み待ち（状況に応じて調整）
    except NoSuchElementException:
        print("❌ もっと見るボタンが見つかりません（おそらくもうない）")
        break
    except ElementClickInterceptedException:
        print("⚠️ ボタンが他の要素に覆われて押せません")
        time.sleep(2)
        continue

# --- 記事のタイトルとリンクを取得して固定リストに保存 ---
article_elements = driver.find_elements(By.XPATH, '//a[contains(@href, "/articles/")]')
articles = []
for a in article_elements:
    href = a.get_attribute("href")
    title = a.text.strip()
    if href and title:
        articles.append((title, href))

# --- 結果格納用 ---
data = {}

# --- 各記事のコメントページへアクセス ---
for index, (title, href) in enumerate(articles):
    print(f"[{index+1}] タイトル: {title}")
    print(f"URL: {href}")

    # コメントページにアクセス
    driver.get(href + "/comments")
    time.sleep(3)  # コメント読み込み待ち

    # コメント要素の取得（data-testid="comment" が安定）
    comments_ = driver.find_elements(By.CLASS_NAME, "hYFULX")
    comments  = [c.text for c in comments_ if len(comments_) > 0] 

    data[href] = {
        "title": title,
        "url": href,
        "comments": comments
    }

driver.quit()

# --- JSONに保存 ---
filename = f"politics_comments_{datetime.datetime.now().strftime('%Y%m%d%H%M%S')}.json"
with open(filename, "w", encoding="utf-8") as f:
    json.dump(data, f, ensure_ascii=False, indent=4)

print(f"\n✅ 完了：{len(data)}記事のコメントを保存 → {filename}")
```
